```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np
```

```python
def show_image(image):
    plt.figure(figsize=(10, 8))
    plt.imshow(image)
    plt.axis('off')
```

```python
completo = Image.open(r'C:\Users\Ceiec06\Documents\GitHub\ARQGAN\other\pruebas\color_complete_1.png')
ruinas = Image.open(r'C:\Users\Ceiec06\Documents\GitHub\ARQGAN\other\pruebas\color_incomplete_1.png')
```

```python
show_image(completo)
```

```python
show_image(ruinas)
```

```python
ruinas = np.array(ruinas)
completo = np.array(completo)
partes_iguales = ruinas != completo

partes_faltantes = np.where(partes_iguales != False, completo, 0)
faltantes_pil = Image.fromarray(partes_faltantes)

filtered = faltantes_pil.filter(ImageFilter.MedianFilter(5))
show_image(filtered)
```

```python
def add_bg(path_img, path_colors, path_background):
    image = Image.open(path_img)
    colors = Image.open(path_colors)
    background = Image.open(path_background)
```
