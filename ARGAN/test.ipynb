```python
import argparse
import os
import tensorflow as tf

from datetime import datetime
from utils import data

from keras_parts import builder

```

```python
model_path = 'saved_models/512x512.pix2pix.color_assisted.t1456789.h5'
log_dir = 'logs/enrique/'
samples_path = 'dataset/temples_ruins/temple_9_ruins_0/'
width = 512
height = 512
```

```python
model = tf.keras.models.load_model(model_path, compile=False)

# builder.get_model('pix2pix', 'color_assisted', 1456789, 'batch')

model.summary()
```

```python
dataset = data.get_simple_dataset(width, height, samples_path + r'\*')
dataset = data.get_dataset(samples_path, 'color_assisted')

name_pos = samples_path.rfind('\\') + 1
folder_name = samples_path[name_pos:]
time = datetime.now().strftime('%Y%m%d-%H%M%S')
log_path = '/'.join([log_dir, model.name, folder_name, time])

writer = tf.summary.create_file_writer(log_path)
with writer.as_default():
    with tf.name_scope('images'):
        for i, x in enumerate(dataset):
            gx = model(x, training=True)
            tf.summary.image('x', tf.squeeze(x, axis=0) * 0.5 + 0.5, step=i)
            tf.summary.image('gx', gx * 0.5 + 0.5, step=i)
```
